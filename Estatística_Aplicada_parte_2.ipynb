# 📦 1. Importando Bibliotecas


```python
import pandas as pd
import kagglehub
import os
```

# 📥 2. Baixando o Dataset do Kaggle


```python
path = kagglehub.dataset_download("hubertsidorowicz/football-players-stats-2024-2025")
print("Path to dataset files:", path)
```

# 📂 3. Carregando o arquivo events.csv


```python
players=pd.read_csv(os.path.join(path, 'players_data-2024_2025.csv'))
```

# 👀 4. Visualizando as Primeiras Informações


```python
players.head()
```

# 📏 5. Verificando o Tamanho do Dataset


```python
print("Número de linhas e colunas:",players.shape)
```

# 🏷️ 6. Explorando as Variáveis


```python
print(players[['Player', 'Nation', 'Pos', 'Squad', 'Comp',
               'Age', 'Born', 'MP', 'Starts', 'Min',
               '90s', 'Gls', 'Ast', 'G+A', 'xG',
                'npxG', 'G-PK', 'Tkl', 'TklW',
               'Blocks', 'Int', 'Tkl+Int', 'Clr', 'Err',
               'PrgP', 'PrgC', 'KP', 'Cmp%',
               'xA', 'PPA', 'GA', 'Saves', 'Save%',
               'CS', 'CS%', 'PKA', 'PKsv', 'Touches',
               'Carries', 'PrgR', 'Mis', 'Dis',
               'CrdY', 'CrdR', 'PKwon', 'PKcon', 'Recov']].dtypes)
```

# 🧠 7. Classificando as Variáveis


| Variável           | Tipo         | Observação |
|--------------------|--------------|------------|
| `Player`           | Qualitativa  | Nome do jogador |
| `Nation`          | Qualitativa  | Nacionalidade do jogador |
| `Position`        | Qualitativa | Posição |
| `Squad`        | Qualitativa |Time do jogador  |
| `Comp`        | Qualitativa |Competição que o jogador joga  |
| `Age`        | Quantitativa |Idade do jogador  |
| `Born`        | Qualitativa |Ano de nascimento do jogador  |
| `MP`        | Quantitativa |Partidas jogadas  |
| `Starts`        |Quantitativa  |Partidas iniciadas  |
| `Min`        |Quantitativa  |Minutos jogados  |
| `90s`        |Quantitativa  |Número de jogos completos (90 min)  |
| `Gls`        |Quantitativa  |Gols marcados  |
| `Ast`        |Quantitativa  |Assistências fornecidas  |
| `G+A`        |Quantitativa  |Gols marcados somados com assistências fornecidas  |
| `xG`        |Quantitativa  |Gols marcados esperados  |
| `npxG`        |Quantitativa  |Gols esperados excluindo pênaltis  |
| `G-PK`        |Quantitativa  |Gols marcados subtraindo pênaltis marcados  |
| `Tkl`        |Quantitativa  |Desarmes tentados|
| `TklW`        |Quantitativa  |Desarmes vencidos |
| `Blocks`        |Quantitativa  |Chutes bloqueados  |
| `Int`        |Quantitativa  |Interceptações  |
| `Tkl+Int`        |Quantitativa  |Desarmes somados com interceptações  |
| `Clr`        |Quantitativa  |Rebatidas de bola para longe da sua área  |
| `Err`        |Quantitativa  |Erros que causaram gols  |
| `PrgP`        |Quantitativa  |Passes progressivos  |
| `PrgC`        |Quantitativa  |Conduzir a bola progressivamente  |
| `KP`        |Quantitativa  |Passes chave (Passes que levam a finalização)  |
| `Cmp%`        |Quantitativa  |Porcentagem dos passes que foram completos|
| `xA`        |Quantitativa |Assistências esperadas |
| `PPA`        |Quantitativa  |Passes na área adversária  |
| `GA`        |Quantitativa  |Gols sofridos (goleiro) |
| `Saves`        |Quantitativa  |Defesas feitas (goleiro) |
| `Save%`        |Quantitativa  |Porcentagem dos chutes que foram defendidos (goleiro) |
| `CS`        |Quantitativa  |Jogos sem sofrer gols (goleiro) |
| `CS%`        |Quantitativa  |Porcentagem do total de jogos em que não sofreu gols (goleiro) |
| `PKA`        |Quantitativa  |Pênaltis cobrados contra (goleiro)   |
| `PKsv`        |Quantitativa  | Pênaltis defendidos (goleiro)  |
| `Touches`        |Quantitativa  |Total de toques na bola  |
| `Carries`        |Quantitativa  |Número de conduções de bola  |
| `PrgR`        |Quantitativa  | Corridas progressivas |
| `Mis`        |Quantitativa  |Erros de domínio de bola  |
| `Dis`        |Quantitativa  |Quanta vezes o jogador foi desarmado  |
| `CrdY`        |Quantitativa  |Cartões amarelos  |
| `CrdR`        |Quantitativa  |Cartões vermelhos  |
| `PKwon`        |Quantitativa  |Pênaltis sofridos  |
| `PKcon`        |Quantitativa  |Pênaltis concedidos  |
| `Recov`        |Quantitativa  |Recuperações de bola  |
